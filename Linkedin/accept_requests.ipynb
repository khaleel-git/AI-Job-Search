```python

url = "https://www.linkedin.com/mynetwork/invitation-manager/received/"
driver.get(url)
```

```python
def scroll_and_accept_connections(driver):
    print("üìú Scrolling to load ALL connection requests...\n")
    
    # Step 1: Infinite scroll to load everything
    last_height = 0
    no_change_count = 0
    
    while no_change_count < 3:
        driver.execute_script("window.scrollBy(0, 800);")
        time.sleep(2)
        
        new_height = driver.execute_script("return document.body.scrollHeight")
        
        if new_height == last_height:
            no_change_count += 1
            print(f"Checking for more... ({no_change_count}/3)")
        else:
            no_change_count = 0
            print(f"Loading more connections... (Height: {new_height})")
        
        last_height = new_height
    
    print("\n‚úÖ Finished loading all connections!\n")
    
    # Step 2: Accept all at once
    Accept_btns = driver.find_elements(By.XPATH, "//button[.//span[text()='Accept']]")
    total = len(Accept_btns)
    
    print(f"üìä Found {total} connection requests to accept\n")
    
    for i, btn in enumerate(Accept_btns, start=1):
        try:
            driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", btn)
            time.sleep(0.3)
            
            print(f"‚úÖ Accepting {i}/{total} ...", end=" ")
            btn.click()
            print("Success!")
            
            wait_time = random.uniform(6.5, 17.3)
            print(f"‚è≥ Waiting {wait_time:.1f}s...")
            time.sleep(wait_time)
            
        except Exception as e:
            print(f"‚úó Failed: {e}")
    
    print(f"\n{'='*60}")
    print(f"üéâ COMPLETED! Accepted {total} connections")
    print(f"{'='*60}")

# Call the function
scroll_and_accept_connections(driver)
```
